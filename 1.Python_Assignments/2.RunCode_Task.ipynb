```python
height = np.random.normal(140, 20, 1000)
plt.figure(dpi=200)
plt.title("Height Distribuion")
plt.xlabel("Height (cm) ")
plt.ylabel("Frequency")
plt.hist(height, bins=30, ec="gold")
plt.show()
```

<h3>Q4. Write Down difference between interpreter and compiler </h3>

<b>Compiler</b> :- Compiler checks the program as a whole and searches for errors. After no error is detected it generates an executable byte code file which on executing gives us the output.

<b>Interpreter</b>:- Interpreter checks the entire code line by line. So if a line contains error it stops executing. It is slower than compiler.

A data structure is a data organization, management, and storage format that is usually chosen for efficient access to data.
